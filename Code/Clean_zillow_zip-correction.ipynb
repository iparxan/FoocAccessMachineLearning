```python
import pandas as pd
```

```python
# Files to load
zillow_zip_csv = "Resources/Zip_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv"

# Read the CSV into a DataFrame
zillow_zip_df = pd.read_csv(zillow_zip_csv)
zillow_zip_df
```

```python
zillow_zip_df.dtypes
```

```python
for column in zillow_zip_df.columns:
    print(f"Column {column}: {zillow_zip_df[column].isnull().sum()} null values and type: {type(column)}")
```

```python
clean_zillow_zip_df= zillow_zip_df[["RegionID","SizeRank","RegionName", "State", "City", "CountyName", "12/31/2015", "12/31/2019"]]
clean_zillow_zip_df
```

```python
ca_clean_zillow_zip_df = clean_zillow_zip_df.loc[clean_zillow_zip_df["State"] == "CA", :]
ca_clean_zillow_zip_df
```

```python
ca_zillow_zip_df = ca_clean_zillow_zip_df.rename(columns={"RegionName": "ZIP"})
ca_zillow_zip_df
```

```python
for column in ca_clean_zillow_zip_df.columns:
    print(f"Column {column}: {ca_clean_zillow_zip_df[column].isnull().sum()} null values and type: {type(column)}")
```

```python
# Files to load
zip_tract_csv = "Resources/ZIP_TRACT_122015.csv"

# Read the CSV into a DataFrame
zip_tract_df = pd.read_csv(zip_tract_csv)
zip_tract_df
```

```python
# Merge the two DataFrames together based on the Dates they share
housing_value_df = pd.merge(ca_zillow_zip_df, zip_tract_df, on="ZIP")
# Merge the two DataFrames together based on the Dates they share
housing_value_df
```

```python
housing_value_df=housing_value_df.drop(columns=["RegionID"],axis=1)
```

```python
housing_value_df= housing_value_df[["TRACT","ZIP", "State", "CountyName", "12/31/2015", "12/31/2019"]]
housing_value_df
```

```python
housing_value_df.count()
```

```python
# drop null rows
no_null_housing_value_df = housing_value_df.dropna(how='any')
no_null_housing_value_df.count()
```

```python
housing_value_clean_df =  no_null_housing_value_df.rename(columns={"TRACT": "CensusTract"})
housing_value_clean_df
```
