```python
housing_price=clean_house_price.filter(['CensusTract','Estimate_Median_value_(dollars)',
       'Margin_of_Error_Median_value_(dollars)'], axis=1)
housing_price
```

```python
#alternative merge
FoodAccess2015HPAll_df=pd.merge(foodAccess15All_df, housing_price, how="left", on=["CensusTract"],validate=None)
FoodAccess2015HPAll_df.head(10)
```

```python
FoodAccess2015HPAll_df=FoodAccess2015HPAll_df.astype({"Estimate_Median_value_(dollars)": float, "Margin_of_Error_Median_value_(dollars)": float},errors='ignore')
FoodAccess2015HPAll_df
```

```python
def missing_values_table(FoodAccess2015HPAll_df):
    mis_val = FoodAccess2015HPAll_df.isnull().sum()
    mis_val_percent = 100 * FoodAccess2015HPAll_df.isnull().sum() / len(FoodAccess2015HPAll_df)
    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
    mis_val_table_ren_columns = mis_val_table.rename(
    columns = {0 : 'Missing Values', 1 : '% of Total Values'})
    mis_val_table_ren_columns = mis_val_table_ren_columns[
        mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
    '% of Total Values', ascending=False).round(1)
    print ("Your selected dataframe has " + str(FoodAccess2015HPAll_df.shape[1]) + " columns.\n"      
        "There are " + str(mis_val_table_ren_columns.shape[0]) +
            " columns that have missing values.")
    return mis_val_table_ren_columns
```

```python
 missing_values_table(FoodAccess2015HPAll_df)
```
